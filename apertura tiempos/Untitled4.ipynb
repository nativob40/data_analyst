# # Importo las librerias

```python
import pandas as pd
import numpy as np
import os
```

## Leo el CSV

```python
df=pd.read_csv(f'{os.getcwd()}/b_12_01_21.csv',
            sep=',',
            header=1,
            index_col=False,)
```

```python
df.isnull().sum()
```

```python
limpio=df.interpolate()
```

```python
limpio
```

```python
limpio.isnull().sum()
```

```python
apertura=pd.DataFrame(limpio,columns=['NumeroCucharaColando','Heat Number','Tiempo_Apertura_1','Tiempo_Apertura_2'])
```

```python
apertura.corr()
```

```python
apertura_pivot = apertura.pivot_table(values=['Tiempo_Apertura_1','Tiempo_Apertura_2'],
                     index=['NumeroCucharaColando','Heat Number'],
                     aggfunc={'Tiempo_Apertura_1': np.max,
                              'Tiempo_Apertura_2': np.max})
```

```python
apertura_pivot[(apertura_pivot['Tiempo_Apertura_1'] != 0) | (apertura_pivot['Tiempo_Apertura_2'] !=0)][1:] # [1:] Filtra "NumeroCucharaColando" != 0
```
