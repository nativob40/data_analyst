## Importo las librerias

```python
import pandas as pd
import numpy as np
import os
```

```python
directorio = os.getcwd()
```

## Leo el CSV de IBA

```python
df=pd.read_csv(f'{directorio}/iba_12_01_21.csv',

            sep=',',

            header=1,

            index_col=False)
```

```python
df.isnull().sum()
```

```python
limpio=df.interpolate() #Interpolo porque son cortes muy pequeños de la señal. Algun error en el archivo
```

```python
limpio.isnull().sum()
```

```python
apertura=pd.DataFrame(limpio,columns=['Indice','NumeroCucharaColando','Heat Number','Tiempo_Apertura_1','Tiempo_Apertura_2','_AVG_Presion_B1_A','_AVG_Presion_B1_B','_AVG_Presion_B2_A','_AVG_Presion_B2_B'])
```

```python
titulos = ['Indice','NumeroCucharaColando','nroColada','Tiempo_Apertura_1','Tiempo_Apertura_2','Cam_A_Arm1','Cam_B_Arm1','Cam_A_Arm2','Cam_B_Arm2']
```

```python
apertura.columns = titulos
```

```python
apertura[(apertura['Tiempo_Apertura_1']>0) | (apertura['Tiempo_Apertura_2']>0)]
```

```python
apertura=apertura[(apertura['Tiempo_Apertura_1']>0) | (apertura['Tiempo_Apertura_2']>0)]
```

```python
tabla_iba = apertura.pivot_table(values=['nroColada','Tiempo_Apertura_1','Tiempo_Apertura_2','Cam_A_Arm1','Cam_B_Arm1','Cam_A_Arm2','Cam_B_Arm2'],
                     index=['Indice'],
                     aggfunc={'nroColada': np.max,
                              'Tiempo_Apertura_1': np.max,
                              'Cam_A_Arm1': np.max,
                              'Cam_B_Arm1': np.max,           
                              'Tiempo_Apertura_2': np.max,
                              'Cam_A_Arm2': np.max,
                              'Cam_B_Arm2': np.max,                              
                             })
```

```python
tabla_iba
```

```python
a_tabla_iba=tabla_iba.to_numpy(dtype=np.float64)
```

### Desplazo un registro hacia arriba la columna del N° de colada

```python
nroColada=[]

for i in range(1,len(a_tabla_iba-1)):
    nroColada.append(a_tabla_iba[i][6])
```

```python
a_nroColada=np.asarray(nroColada)
```

```python
a_tabla_iba=np.delete(a_tabla_iba,6,1) #elimino la columna del N°6 "nroColada" en el array "a_tabla_iba"
```

### Genero los 2 DF

```python
colada=pd.DataFrame(nroColada,columns=['nroColada'])
```

```python
tabla_iba=pd.DataFrame(a_tabla_iba,columns=['Cam_A_Arm1','Cam_A_Arm2','Cam_B_Arm1','Cam_B_Arm2','Tiempo_Apertura_1','Tiempo_Apertura_2'])
```

```python
tabla_iba=pd.concat((tabla_iba,colada),axis=1)
```

```python
tabla_iba.dropna(inplace=True) #Elimino el ultimo registro de la columna "nroColada" por ser NaN
```

# Leo el XLS de SIO

```python
tabla_sio=pd.read_excel(f'{directorio}/sio_12_01_21.xlsx',sheet_name='Hoja1')
```

```python
tabla_sio=pd.DataFrame(tabla_sio,columns=['nroColada','cuchara_nro'])
```

## Uno las Tablas

```python
tabla_final = pd.merge(tabla_sio, tabla_iba, on='nroColada')
```
