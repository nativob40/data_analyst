```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv(r'C:\Users\ddevia\Desktop\Consumo RAM\tabla_Analisis_ram_molde_L3.csv',sep=',')
```

```python
inf_der = pd.DataFrame(df, columns=['TimeSeconds',
                                    's3_tcs_ram_actcurbsls_C1074135149',
                                    's3_tcs_ram_actcurbsrs_C1074135150',
                                    's3_tcs_ram_actcurtsls_C1074135151',
                                    's3_tcs_ram_actcurtsrs_C1074135152',
                                    's3_an2l_ramzustand_alt_C1075314736'])
```

# Filtro solo cuando esta en la FASE 2 --- Paso 176

```python
inf_der=inf_der[inf_der['s3_an2l_ramzustand_alt_C1075314736']==176]
```

# Armo el DataFrame --- Direccion

### Miro que este en el paso 176 y comparo la referencia con la posicion actual para ver el sentido

```python
fecha=[]
sentido=[]
posicion=[]

for s in range(1,len(df)):
    
    if df.loc[s,'s3_an2l_ramzustand_alt_C1075314736'] == 176: 
    
        if df.loc[s,'s3_an2e_ramrefautocoldwidthhmi_C2185'] < df.loc[s,'s3_an2l_ramactwidthboc_C1547']:
            fecha.append(df.loc[s,'TimeSeconds'])
            sentido.append('achicando')

        elif df.loc[s,'s3_an2e_ramrefautocoldwidthhmi_C2185'] > df.loc[s,'s3_an2l_ramactwidthboc_C1547']:
            fecha.append(df.loc[s,'TimeSeconds'])
            sentido.append('agrandando')

        elif df.loc[s,'s3_an2e_ramrefautocoldwidthhmi_C2185'] == df.loc[s,'s3_an2l_ramactwidthboc_C1547']:
            fecha.append(df.loc[s,'TimeSeconds'])
            sentido.append('detenido')
```

```python
direccion={'fecha':fecha,'sentido':sentido}

df_direccion = pd.DataFrame(data=direccion)
```

# Armo las diferentes categorias

```python
fecha=[]
rango=[]

for s in range(1,len(df)):
    
    if df.loc[s,'s3_an2l_ramzustand_alt_C1075314736'] == 176: 

            decimal,entero=math.modf(df.loc[s,'s3_an2l_ramactwidthboc_C1547']/100)
            rango.append(int(entero*100))
            fecha.append(df.loc[s,'TimeSeconds'])
            
```

# Armo el DataFrame --- Rango

```python
rangos={'fecha':fecha,'rango':rango}

df_rango = pd.DataFrame(data=rangos)
```

```python
consumos=df_rango.set_index('fecha').join(df_direccion.set_index('fecha')).join(inf_der.set_index('TimeSeconds'), on='fecha')
```

```python
consumos.boxplot(column=['s3_tcs_ram_actcurtsls_C1074135151','s3_tcs_ram_actcurtsrs_C1074135152',
                         's3_tcs_ram_actcurbsls_C1074135149','s3_tcs_ram_actcurbsrs_C1074135150'],by='rango',
                 layout=(2, 2),
                 showmeans=True, 
                 showbox=True,
                 figsize=(20,15),
                 boxprops={'color':'blue'},
                 medianprops={'color':'red'},
                 whiskerprops={'color':'blue'},
                 meanprops={'marker':'o','mec':'black','mfc':'grey'})
plt.show()
```
