```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import os
```

```python
carpetas=os.getcwd()+'\\datos\\'
archivos=os.listdir(carpetas)
```

```python
df = pd.read_csv(carpetas+archivos[0],sep=',')
```

```python
inf_der = pd.DataFrame(df, columns=['TimeSeconds',
                                    's4_tcs_ram_actcurbsls_C1074200685',
                                    's4_tcs_ram_actcurbsrs_C1074200686',
                                    's4_tcs_ram_actcurtsls_C1074200687',
                                    's4_tcs_ram_actcurtsrs_C1074200688',
                                    's4_an2l_ramzustand_alt_C1075380272'])
```

# Filtro solo cuando esta en la FASE 2 --- Paso 176

```python
inf_der=inf_der[inf_der['s4_an2l_ramzustand_alt_C1075380272']==176]
```

# Armo el DataFrame --- Direccion

### Miro que este en el paso 176 y comparo la referencia con la posicion actual para ver el sentido

```python
fecha=[]
sentido=[]
posicion=[]

for s in range(1,len(df)):
    
    if df.loc[s,'s4_an2l_ramzustand_alt_C1075380272'] == 176: 
    
        if df.loc[s,'CC2.H.AN2E_RamRefAutoColdWidthHmi'] < df.loc[s,'s4_an2l_ramactwidthboc_C1611']:
            fecha.append(df.loc[s,'TimeSeconds'])
            sentido.append('achicando')

        elif df.loc[s,'CC2.H.AN2E_RamRefAutoColdWidthHmi'] > df.loc[s,'s4_an2l_ramactwidthboc_C1611']:
            fecha.append(df.loc[s,'TimeSeconds'])
            sentido.append('agrandando')

        elif df.loc[s,'CC2.H.AN2E_RamRefAutoColdWidthHmi'] == df.loc[s,'s4_an2l_ramactwidthboc_C1611']:
            fecha.append(df.loc[s,'TimeSeconds'])
            sentido.append('detenido')
```

```python
direccion={'fecha':fecha,'sentido':sentido}

df_direccion = pd.DataFrame(data=direccion)
```

# Armo las diferentes categorias

```python
fecha=[]
rango=[]

for s in range(1,len(df)):
    
    if df.loc[s,'s4_an2l_ramzustand_alt_C1075380272'] == 176: 

            fecha.append(df.loc[s,'TimeSeconds'])
            decimal,entero=math.modf(df.loc[s,'s4_an2l_ramactwidthboc_C1611']/100)
            rango.append(int(entero*100))
            
```

# Armo el DataFrame --- Rango

```python
rangos={'fecha':fecha,'rango':rango}

df_rango = pd.DataFrame(data=rangos)
```

```python
consumos=df_rango.set_index('fecha').join(df_direccion.set_index('fecha')).join(inf_der.set_index('TimeSeconds'), on='fecha')
```

```python
num_molde=3

fecha = pd.to_datetime(df['TimeSeconds'])
fecha = fecha.dt.strftime('%d/%m/%Y') 

periodo=f"{fecha[min(fecha.index)]} al {fecha[max(fecha.index)]}" 
```

```python
ax=consumos.boxplot(column=['s4_tcs_ram_actcurtsls_C1074200687','s4_tcs_ram_actcurtsrs_C1074200688',
                         's4_tcs_ram_actcurbsls_C1074200685','s4_tcs_ram_actcurbsrs_C1074200686'],by='rango',
                 layout=(2, 2),
                 showmeans=True, 
                 showbox=True,
                 figsize=(20,15),
                 boxprops={'color':'blue'},
                 medianprops={'color':'red'},
                 whiskerprops={'color':'blue'},
                 meanprops={'marker':'o','mec':'black','mfc':'grey'})

fig = ax[0][0].get_figure()
fig.suptitle(f'\n\nConsumos Motores RAM L#4 --- Molde NÂ°: {num_molde} --- Periodo: {periodo}')
ax[0][0].set_title("Sup_Izq")
ax[0][1].set_title("Sup_Der")
ax[1][0].set_title("Inf_Izq")
ax[1][1].set_title("Inf_Der")

ax[0][0].set_ylabel("Corriente")
#ax[0][1].set_ylabel("Corriente")
ax[1][0].set_ylabel("Corriente")
#ax[1][1].set_ylabel("Corriente")

#ax[0][0].set_xlabel("Ancho")
#ax[0][1].set_xlabel("Ancho")
ax[1][0].set_xlabel("Ancho")
ax[1][1].set_xlabel("Ancho")

#plt.show()

plt.savefig('Consumos Motores RAM L4.png',facecolor='w')
```
